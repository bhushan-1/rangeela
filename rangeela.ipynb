```python
import tkinter as tk
from tkinter import *
from PIL import ImageTk,Image
from tkinter import messagebox
import random 

# frames function
def show_frame(frame):
    frame.tkraise()
    
w=tk.Tk()
w.geometry('400x550')
w.resizable(0,0)
w.title("RANGEELA")


#######################################

  
# list of possible colour. 
colours = ['Red','Blue','Green','Pink','Black', 
           'Yellow','Orange','White','Purple','Brown'] 
score = 0
  
# the game time left, initially 30 seconds. 
timeleft = 30
  
# function that will start the game. 
def startGame(event): 
      
    if timeleft == 30: 
          
        # start the countdown timer. 
        countdown() 
          
    # run the function to 
    # choose the next colour. 
    nextColour() 
  
# Function to choose and 
# display the next colour. 
def nextColour(): 
  
    # use the globally declared 'score' 
    # and 'play' variables above. 
    global score 
    global timeleft 
  
    # if a game is currently in play 
    if timeleft > 0: 
  
        # make the text entry box active. 
        e.focus_set() 
  
        # if the colour typed is equal 
        # to the colour of the text 
        if e.get().lower() == colours[1].lower(): 
              
            score += 1
  
        # clear the text entry box. 
        e.delete(0, tk.END) 
          
        random.shuffle(colours) 
          
        # change the colour to type, by changing the 
        # text _and_ the colour to a random colour value 
        label.config(fg = str(colours[1]), text = str(colours[0])) 
          
        # update the score. 
        scoreLabel.config(text = "      Score: " + str(score)) 
  
  
# Countdown timer function  
def countdown(): 
  
    global timeleft 
  
    # if a game is in play 
    if timeleft > 0: 
  
        # decrement the timer. 
        timeleft -= 1
          
        # update the time left label 
        timeLabel.config(text = "Time left: "
                               + str(timeleft)) 
                                 
        # run the function again after 1 second. 
        timeLabel.after(1000, countdown) 
#######################################


frame1=tk.Frame(w)
frame2=tk.Frame(w)
frame3=tk.Frame(w)


for frame in (frame1,frame2,frame3):
    frame.place(x=0,y=0)
    
###############


r = ImageTk.PhotoImage(file="a.jpg")
b= Label(frame1,image = r)
b.pack()


value= StringVar()




def get_button():
    name= value.get()
    show_frame(frame2)
  
frame1_title =tk.Label(frame1,text="ENTER THE PLAYER NAME",bg="#ffd47f", font= (45))
frame1_title.place(x=135,y=250)



frame1_e1= tk.Entry(frame1,bd=5, justify= "center", font= ("arial", 16,"bold"), textvariable= value)
frame1_e1.place(x=140,y=315)
frame1_btn= tk.Button(frame1,text= "Submit",font= 40,command= get_button)
frame1_btn.place(x=220,y=370)

###############


re = ImageTk.PhotoImage(file="bgall.jpg")
bf= Label(frame2,image = re)
bf.pack()




frame2_title =tk.Label(frame2,text="Welcome!",bg="#ffd47f", font= ("ms serif", 30,"bold"))
frame2_title.place(x=110,y=50)
frame2_title =tk.Label(frame2,textvariable= value,bg="#ffd47f", font= ("ms serif", 30,"bold") )
frame2_title.place(x=135,y=100)

#pop up

def pop():
    try:
        int(frame2_e1.get())
        show_frame(frame3)
        
    except: 
        messagebox.showwarning('pop up','Enter in digits')
        



frame2_title =tk.Label(frame2,text="ENTER YOUR AGE",bg="#ffd47f", font= ("times roman", 20,"bold"))
frame2_title.place(x=80,y=290)
frame2_e1= tk.Entry(frame2,bd=5, justify= "center", font= ("arial", 20,"bold"),width="5")
frame2_e1.place(x=155,y=340)

frame2_btn= tk.Button(frame2,text= "Submit",font= 10,command=pop,width="6")
frame2_btn.place(x=160,y=400)

###############




'''

frame3_btn= tk.Button(frame2,text= "END",font= 40,command=w.destroy)
frame3_btn.place(x=190,y=500)
'''


###################################################################
# create a GUI window 
frame3= tk.Tk() 
  
# set the title 
frame3.title("RANGEELA") 
  
# set the size 
frame3.geometry("400x550") 
frame3.resizable(0,0)
 
# add an instructions label 
instructions = tk.Label(frame3, text = "Type in the colour "
                        "of the words, and not the word text!", 
                                      font = ('algerian', 12)) 
instructions.place(x=20,y=20)
  
# add a score label 
scoreLabel = tk.Label(frame3, text = "Press enter to start", 
                                      font = ('times roman', 16)) 
scoreLabel.place(x=130,y=50)
  
# add a time left label 
timeLabel = tk.Label(frame3, text = "Time left: " +
              str(timeleft), font = ('times roman', 16)) 
                
timeLabel.place(x=145,y=80)
  
# add a label for displaying the colours 
label = tk.Label(frame3, font = ('times roman', 60)) 
label.place(x=115,y=140)

def end():
    w.destroy()
    frame3.destroy()


frame3_btn= tk.Button(frame3,text= "END GAME",font= 40,command=lambda:end())
frame3_btn.place(x=160,y=400)
  
# add a text entry box for 
# typing in colours 
e = tk.Entry(frame3,font= ("arial", 9,"bold"),justify= "center") 
  
# run the startGame function  
# when the enter key is pressed 
frame3.bind('<Return>', startGame) 
e.place(x=155,y=240) 
  
# set focus on the entry box 
e.focus_set() 
  





show_frame(frame1)

w.mainloop()
```
